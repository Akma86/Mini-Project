```python
import numpy as np
import mediapipe as mp
from cvzone.HandTrackingModule import HandDetector
import cv2 as cv
import time
import math
import pyautogui
import pygetwindow as gw
```

```python
def is_palm_facing(hand_landmarks, tolerance = 0.5):
    visible = True
    finger_landmarks = [
        (0,4),
        (5,8)
    ]
    for base, tip in finger_landmarks:
        base_z = hand_landmarks[base][2]
        tip_z = hand_landmarks[tip][2]

        if tip_z >= base_z - tolerance:
            visible = False
            break
    return visible
```

```python
pTime = 0
previous_x , previous_y = 0 , 0
smoothing_factor = 0.5 
screen_width, screen_height = pyautogui.size()
dual_screen_height = screen_height * 2
video = cv.VideoCapture(0)
is_dragging = False
initial_position = (0, 0)
video.set(cv.CAP_PROP_FRAME_WIDTH, screen_width)
video.set(cv.CAP_PROP_FRAME_HEIGHT, screen_height)

detector=HandDetector(detectionCon=0.8, maxHands=2)
while True:
    ret,frame = video.read()
    frame = cv.flip(frame,1)

    #define fps
    cTime = time.time()
    fps = 1/(cTime - pTime)
    pTime=cTime
    
    #find hands
    #hands disini berisi koordinat dari setiap landmark
    hands, image = detector.findHands(frame)
    if hands:
        for hand in hands:   
            hand_landmarks = hand['lmList'] 
            hand_Types = hand['type'] 
            bbox = hand['bbox'] # bounding box dari tangan 
            centerPoint = hand['center'] # koordinat tengah tangan
            if is_palm_facing(hand_landmarks):
                fingerUp = detector.fingersUp(hand)
                if sum(fingerUp) == 0:
                    pyautogui.mouseUp()
                    cv.putText(image,'Normal',(40,40),cv.FONT_HERSHEY_COMPLEX_SMALL,1,(0,188,0),1)
                elif fingerUp == [0,1,0,0,0]:
                    index_finger_x, index_finger_y = hand_landmarks[8][0], hand_landmarks[8][1]
                    frame_height, frame_width, _ = frame.shape
                    screen_x = int(index_finger_x * screen_width / frame_width)
                    screen_y = int(index_finger_y * screen_height / frame_height)
                    pyautogui.moveTo(screen_x, screen_y)
                    cv.putText(image,'Move',(40,40),cv.FONT_HERSHEY_COMPLEX_SMALL,1,(0,188,0),1)
                elif fingerUp == [0,1,1,0,0]:
                    pyautogui.click()
                    cv.putText(image,'Click',(40,40),cv.FONT_HERSHEY_COMPLEX_SMALL,1,(0,188,0),1)
                    time.sleep(0.5)
                elif fingerUp == [1, 1, 1, 0, 0]:
                    index_finger_x, index_finger_y = hand_landmarks[8][0], hand_landmarks[8][1]
                    frame_height, frame_width, _ = frame.shape
                    screen_x = int(index_finger_x * screen_width / frame_width)
                    screen_y = int(index_finger_y * screen_height / frame_height)

                    if not is_dragging:
                        pyautogui.mouseDown()  # Mulai drag
                        is_dragging = True
                        pyautogui.dragTo(screen_x, screen_y, duration=0.1)   
                    else:
                        pyautogui.moveTo(screen_x, screen_y)  # Lanjutkan drag
                    cv.putText(image, 'Dragging', (40, 40), cv.FONT_HERSHEY_COMPLEX_SMALL, 1, (0, 188, 0), 1)


    #show image
    cv.putText(image,f'fps :{str(int(fps))}',(40,40),cv.FONT_HERSHEY_COMPLEX_SMALL,1,(0,188,0),1)
    cv.imshow("Image",image) 
    if cv.waitKey(1) & 0xFF==ord(' '):
        break
    
video.release()
cv.destroyAllWindows()
```
